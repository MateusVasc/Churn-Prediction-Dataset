```python
df.info()
```

```python
df.isnull().sum()
```

```python
freq = df['TotalCharges'].value_counts() 
print(freq)
```

```python
df = df.drop(df.loc[df['TotalCharges'] == ' '].index)
df.info()
```

```python
df["TotalCharges"] = df["TotalCharges"].astype(float)
df.info()
```
